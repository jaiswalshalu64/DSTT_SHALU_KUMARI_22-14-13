````python
#5 Drop all rows with N/A,NA,na values in Num_Bedrooms
data['NUM_BEDROOMS'] = data['NUM_BEDROOMS'].replace(['N/A', 'na', 'NA'], None) # it replace N/A,na,NA with none boolean
data.dropna(subset=['NUM_BEDROOMS'], inplace=True) # it drop down all rows of column Num_Bedrooms which is having boolean vaue=true 
````

````python
#6 Replace the NaN and String entries in Num_Bath with previous entry
data['NUM_BATH'] = pd.to_numeric(data['NUM_BATH'], errors='coerce') # it convert all numeric string to float and string to nan from column Num_Bath
data['NUM_BATH'] = data['NUM_BATH'].fillna(method='ffill') # it replace nan entry to previous entry
````

````python
data
````

````python
#7 Replace the empty values in “SQ_FT” with the mean of the all entries.
data['SQ_FT'] = pd.to_numeric(data['SQ_FT'], errors='coerce') # it convert all numeric string to float and string to nan from column SQ_FT
mean=data['SQ_FT'].mean() # it calculate mean of column SQ_FT
data['SQ_FT'].fillna(mean, inplace=True) # it replace all null value with mean of SQ_FT
````
